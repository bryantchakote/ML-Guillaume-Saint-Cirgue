```python
fourier = fftpack.fft(y)
power = np.abs(fourier)
frequences = fftpack.fftfreq(y.size)
plt.plot(np.abs(frequences), power)
```

# CHAP 21. SKLearn - Model selection - train_test_split - Cross validation - GridSearch

## 1) Trainset / Testset - train_test_split

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

X = iris.data
y = iris.target

print(X.shape)
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=.8)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=5)

print("Train set:", X_train.shape)
print("Test set:", X_test.shape)
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(121)
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, alpha=.8)
plt.title("Train set")

plt.subplot(122)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, alpha=.8)
plt.title("Test set")
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors = 1)
model.fit(X_train, y_train)
print("Train score:", model.score(X_train, y_train))
print("Test score:", model.score(X_test, y_test))
```

## 2) Validation set - Amliorer le modle

```python
cross_val_score(KNeighborsClassifier(), X_train, y_train, cv=5, scoring="accuracy").mean() # cv: nbre de splits | scoring: metrique
```

# CHAP 21_2. Mtriques de rgression en data science

```python
# Erreur quadratique moyenne
# MSE (Mean Squared Error)
# moyenne des carrs des erreurs

# Erreur absolue moyenne
# MAE (Mean Absolute Error)
# moyenne des valeurs absolues des erreurs
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import *
```

```python
y = np.array([1, 2])
y_pred = np.array([1, 2])
```

```python
print("MAE:", mean_absolute_error(y, y_pred))
print("MSE:", mean_squared_error(y, y_pred))
```

```python
y_pred = np.array([5, 2])

print("MAE:", mean_absolute_error(y, y_pred))
print("MSE:", mean_squared_error(y, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y, y_pred))) # RootMSE
```

```python
# La MSE pnalise beaucoup les GRANDES ERREURS que la MAE

# MSE -> importance linaire de l'erreur avec son amplitude
#     -> accorder une grande importance aux grandes erreurs

# MAE -> importance exponentielle de l'erreur avec son amplitude
#     -> le dataset contient des valeurs aberrantes (outliers)
```

```python
# Une autre mtrique moins sensible aux grandes erreurs (Median Absolute Error)
y = np.array([1, 2, 2, 2, 3, 5])
y_pred = np.array([1, 2, 2, 5, 7, 1000])

print("MAE:", mean_absolute_error(y, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y, y_pred)))
print("MedAE:", median_absolute_error(y, y_pred))
```

```python
# Puisque chaque mesure prsente des avantages et des inconvenients
# Au final ... utilisez les toutes!
# Permet de mieux comprendre les donnes, leur distribution, etc
```
